# Scale the Data

```python
# Split the data to select out our dependent variable and scale the variable in X to the same scale
X = df.drop(['diagnosis'], axis=1)
display(X.shape), display(X.head())
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = pd.DataFrame(X)
X_scaled = scaler.fit_transform(X)
X_scaled
```

# MODEL SELECTION

```python
X.shape
```

```python
y = df['diagnosis']
y.shape
```

```python
# ataain a validation set from the data
np.random.seed(1234)
val_set = X.iloc[0:100,]
val_set
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(256)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.shape, X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(criterion='entropy')
tree_train = tree.fit(X_train, y_train)
tree_test = tree.fit(X_test,y_test)
```

```python
tree_pred = tree.predict(X_test)
```

```python
from sklearn.metrics import fbeta_score, accuracy_score,mean_squared_error, confusion_matrix, ConfusionMatrixDisplay
```

```python
AC = accuracy_score(y_test, tree_pred)
FB = fbeta_score(y_test, tree_pred,average='binary',beta=0.5)
MSE = mean_squared_error(y_test, tree_pred)
print("==== DECISSION TREE ====")
print(f"Accuracy of the model is: {AC}\nf1Beta Score(Considers both Recall and Precission) is: {FB}\nThe Mean Squared Error is: {MSE}")

print(f"\n\nperformance on training set: {tree.score(X_train,y_train)}")
print(f"\n\nperformance on testing set: {tree.score(X_test,y_test)}")
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators=200, criterion='entropy', random_state=42)
RF.fit(X_train,y_train)
RF_pred = RF.predict(X_test)
AC = accuracy_score(y_test, RF_pred)
FB = fbeta_score(y_test, RF_pred,average='binary',beta=0.5)
MSE = mean_squared_error(y_test, RF_pred)
print("==== RANDOM FOREST ====")

print(f"Accuracy of the model is: {AC}\nf1Beta Score(Considers both Recall and Precission) is: {FB}\nThe Mean Squared Error is: {MSE}")
#print(f"\n\nperformance on testing set: {test.score(X_test,y_test)}")
```

```python
# MODEL PERFROMANCE ON TRAINING AND TESTING DATA
train = RF.fit(X_train,y_train)
test = RF.fit(X_test,y_test)

print(f"RANDOM FOREST\nPerformance on training set: {train.score(X_train,y_train)}\nPerformance on testing set: {test.score(X_test,y_test)}")
print(f"Accuracy is : {accuracy_score(y_test, RF.predict(X_test))}")
```
